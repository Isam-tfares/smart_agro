```python
from datetime import datetime

def get_data_of_month_year(data, month, year,column):
    filtered_data = []
    for entry in data:
        timestamp = datetime.strptime(entry['timestamp'], '%Y-%m-%d %H:%M:%S')
        if timestamp.month == month and timestamp.year == year:
            if column=="all":
                filtered_data.append(entry)
            else:
                filtered_data.append({'timestamp':entry['timestamp'],column:entry[column]})
    return filtered_data

get_data_of_month_year(data,5,2024,'Speed')
```

```python
from datetime import datetime

def get_data_of_day_month_year(data, day, month, year):
    filtered_data = []
    for entry in data:
        timestamp = datetime.strptime(entry['timestamp'], '%Y-%m-%d %H:%M:%S')
        if timestamp.day == day and timestamp.month == month and timestamp.year == year:
            filtered_data.append(entry)
    return filtered_data

get_data_of_day_month_year(data,3,5,2024)
```
